```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("E:/ANTT Learning/IRIS.csv")
df.head()
```

```python
import matplotlib.pyplot as plt 
```

```python

plt.scatter(df['petal_length'],df['petal_width'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.petal_length,df1['petal_width'],color='blue')
plt.scatter(df2.petal_length,df2['petal_width'],color='red')
plt.scatter(df3.petal_length,df3['petal_width'],color='green')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df.petal_width)
df['petal_width'] = scaler.transform(df.petal_width)
df

scaler.fit(df.petal_length)
df['petal_length'] = scaler.transform(df.petal_length)
df
```
