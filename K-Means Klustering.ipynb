```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("E:/ANTT Learning/IRIS.csv")
df.head()
```

```python
import matplotlib.pyplot as plt 
```

```python

plt.scatter(df['petal_length'],df['petal_width'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.petal_length, df1['petal_width'], color='blue', label='Cluster 0')
plt.scatter(df2.petal_length, df2['petal_width'], color='red', label='Cluster 1')
plt.scatter(df3.petal_length, df3['petal_width'], color='green', label='Cluster 2')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['petal_width', 'petal_length']])  
df[['petal_width', 'petal_length']] = scaler.transform(df[['petal_width', 'petal_length']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.petal_length, df1['petal_width'], color='blue', label='Cluster 0')
plt.scatter(df2.petal_length, df2['petal_width'], color='red', label='Cluster 1')
plt.scatter(df3.petal_length, df3['petal_width'], color='green', label='Cluster 2')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color='purple',marker='*',label='centroid')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['petal_width','petal_length']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of sqaure error')
plt.plot(k_rng,sse)
```
